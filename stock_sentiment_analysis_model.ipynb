```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.tail()
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
#Removing punctuations

data = train.iloc[:, 2:]
data.replace("[^a-zA-Z]", " ", regex = True, inplace = True)

#Renaming column names for ease of access

new_index = [str(i) for i in range(25)]
data.columns = new_index
data.head()
```

```python
#Converting headlines to lowercase

for index in new_index:
    data[index] = data[index].str.lower()
data.head(1)
```

```python
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, :]))
```

```python
headlines[1]
```

```python
headlines_copy = headlines.copy()
```

## Prediction of Stock sentiment using Machine Learning

```python
### CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
## Implement Bag of words

cv = CountVectorizer(ngram_range=(2,2))
traindataset = cv.fit_transform(headlines)

pickle.dump(cv, open('transform.pkl', 'wb'))
```

```python
traindataset[0]
```

```python
# Implement RandomForest Classifier

randomClassifier = RandomForestClassifier(n_estimators=200, criterion = 'entropy')
randomClassifier.fit(traindataset, train['Label'])

filename = 'stock_sentiment_analysis_model.pkl'
pickle.dump(randomClassifier, open(filename, 'wb'))
```

```python
# Predict the test set results

test_set = []
for row in range(0, len(test.index)):
    test_set.append(' '.join(str(x) for x in test.iloc[row, 2:]))
test_dataset = cv.transform(test_set)
predictions = randomClassifier.predict(test_dataset)
```

```python
## Import library to check accuracy

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
matrix = confusion_matrix(test['Label'], predictions)
print(matrix)
score = accuracy_score(test['Label'], predictions)
print(score)
report = classification_report(test['Label'], predictions)
print(report)
```
